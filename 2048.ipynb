```python
import random
from IPython.display import clear_output

board = []

#creating the board
def start_game():
    for i in range(4):
        board.append([0]*4)
        
def show_board():
    for row in board:
        row_str = ''
        for element in row:
            row_str+=str(element)+'\t'
        print(row_str)
    return

#makes it so board resets
def reset_board():
        board.clear()
        
# in order to add a 2 whenever you move
def add_2(board):
    r = random.randint(0,3)
    c = random.randint(0,3)
    
    while board[r][c] != 0:
        r = random.randint(0,3)
        c = random.randint(0,3)
        
    board[r][c] = 2
#Excrutiating code in order to move the numbers around    
def move(board):
    position = []
    while position not in ['w','a','s','d']:
        position = input('Please choose "w" for up "a" for left "s" for down and "d" for right').lower()
        if position == 'w':
            moved = False
            #To check if the board actually moved
            for j in range(2):
                #So it loops correctly
                for i in range(3):
                    #makes it go 4 times in order combine each
                    if board[i][0] == board[i+1][0] or board[i][0] == 0:
                        board[i][0] += board[i+1][0]
                        board[i+1][0] = 0
                        if board[i][0] != 0 and board[i+1][0] == 0:
                            moved = True
                    if board[i][1] == board[i+1][1] or board[i][1] == 0:
                        board[i][1] += board[i+1][1]
                        board[i+1][1] = 0
                        if board[i][1] != 0 and board[i+1][1] == 0:
                            moved = True
                    if board[i][2] == board[i+1][2] or board[i][2] == 0:
                        board[i][2] += board[i+1][2]
                        board[i+1][2] = 0
                        if board[i][2] != 0 and board[i+1][2] == 0:
                            moved = True
                    if board[i][3] == board[i+1][3] or board[i][3] == 0:
                        board[i][3] += board[i+1][3]
                        board[i+1][3] = 0
                        if board[i][3] != 0 and board[i+1][3] == 0:
                            moved = True                
            #check if there are outliers, ex 0008 -> 8000 not 0800 because of 3 loops
            if board[0][0] == 0 and board[1][0] != 0:  
                board[0][0] += board[1][0]
                board[1][0] = 0
                moved = True
            if board[0][1] == 0 and board[1][1] != 0:  
                board[0][1] += board[1][1]
                board[1][1] = 0
                moved = True
            if board[0][2] == 0 and board[1][2] != 0:  
                board[0][2] += board[1][2]
                board[1][2] = 0
                moved = True
            if board[0][3] == 0 and board[1][3] != 0:  
                board[0][3] += board[1][3]
                board[1][3] = 0
                moved = True
            #If nothing moved, then ask for a different position
            if moved == False:
                print('Please choose a different direction')
                move(board)
                continue
            if moved == True:
                continue

        #rest is basically copy and paste        
        elif position == 's':
            moved = False
            for k in range(2):
                for i in reversed(range(3)):
                    if board[i+1][0] == board[i][0] or board[i+1][0] == 0:
                        board[i+1][0] += board[i][0]
                        board[i][0] = 0
                        if board[i+1][0] != 0 and board[i][0] == 0:
                            moved = True
                    if board[i+1][1] == board[i][1] or board[i+1][1] == 0:
                        board[i+1][1] += board[i][1]
                        board[i][1] = 0
                        if board[i+1][1] != 0 and board[i][1] == 0:
                            moved = True
                    if board[i+1][2] == board[i][2] or board[i+1][2] == 0:
                        board[i+1][2] += board[i][2]
                        board[i][2] = 0
                        if board[i+1][2] != 0 and board[i][2] == 0:
                            moved = True
                    if board[i+1][3] == board[i][3] or board[i+1][3] == 0:
                        board[i+1][3] += board[i][3]
                        board[i][3] = 0
                        if board[i+1][3] != 0 and board[i][3] == 0:
                            moved = True
                            
            if board[3][0] == 0 and board[2][0] != 0:  
                board[3][0] += board[2][0]
                board[2][0] = 0
                moved = True
            if board[3][1] == 0 and board[2][1] != 0:  
                board[3][1] += board[2][1]
                board[2][1] = 0
                moved = True
            if board[3][2] == 0 and board[2][2] != 0:  
                board[3][2] += board[2][2]
                board[2][2] = 0
                moved = True
            if board[3][3] == 0 and board[2][3] != 0:  
                board[3][3] += board[2][3]
                board[2][3] = 0
                moved = True                
            if moved == False:
                print('Please choose a different direction')
                move(board)
                continue
            if moved == True:
                continue

                            
        elif position == 'a':
            moved = False
            for k in range(2):
                for i in range(3):
                    if board[0][i] == board[0][i+1] or board[0][i] == 0:
                        board[0][i] += board[0][i+1]
                        board[0][i+1] = 0
                        if board[0][i] != 0 and board[0][i+1] == 0:
                            moved = True
                    if board[1][i] == board[1][i+1] or board[1][i] == 0:
                        board[1][i] += board[1][i+1]
                        board[1][i+1] = 0
                        if board[1][i] != 0 and board[1][i+1] == 0:
                            moved = True
                    if board[2][i] == board[2][i+1] or board[2][i] == 0:
                        board[2][i] += board[2][i+1]
                        board[2][i+1] = 0
                        if board[2][i] != 0 and board[2][i+1] == 0:
                            moved = True
                    if board[3][i] == board[3][i+1] or board[3][i] == 0:
                        board[3][i] += board[3][i+1]
                        board[3][i+1] = 0
                        if board[3][i] != 0 and board[3][i+1] == 0:
                            moved = True          
            if board[0][0] == 0 and board[0][1] != 0:  
                board[0][0] += board[0][1]
                board[0][1] = 0
                moved = True
            if board[1][0] == 0 and board[1][1] != 0:  
                board[1][0] += board[1][1]
                board[1][1] = 0
                moved = True
            if board[2][0] == 0 and board[2][1] != 0:  
                board[2][0] += board[2][1]
                board[2][1] = 0
                moved = True
            if board[3][0] == 0 and board[3][1] != 0:  
                board[3][0] += board[3][1]
                board[3][1] = 0
                moved = True                  
            if moved == False:
                print('Please choose a different direction')
                move(board)
                continue
            if moved == True:
                continue
                
        elif position == 'd':
            moved = False
            for k in range(2):
                for i in reversed(range(3)):
                    if board[0][i+1] == board[0][i] or board[0][i+1] == 0:
                        board[0][i+1] += board[0][i]
                        board[0][i] = 0
                        if board[0][i+1] != 0 and board[0][i] == 0:
                            moved = True 
                    if board[1][i+1] == board[1][i] or board[1][i+1] == 0:
                        board[1][i+1] += board[1][i]
                        board[1][i] = 0
                        if board[1][i+1] != 0 and board[1][i] == 0:
                            moved = True 
                    if board[2][i+1] == board[2][i] or board[2][i+1] == 0:
                        board[2][i+1] += board[2][i]
                        board[2][i] = 0
                        if board[2][i+1] != 0 and board[2][i] == 0:
                            moved = True 
                    if board[3][i+1] == board[3][i] or board[3][i+1] == 0:
                        board[3][i+1] += board[3][i]
                        board[3][i] = 0
                        if board[3][i+1] != 0 and board[3][i] == 0:
                            moved = True  
            if board[0][3] == 0 and board[0][2] != 0:  
                board[0][3] += board[0][2]
                board[0][2] = 0
                moved = True
            if board[1][3] == 0 and board[1][2] != 0:  
                board[1][3] += board[1][2]
                board[1][2] = 0
                moved = True
            if board[2][3] == 0 and board[2][2] != 0:  
                board[2][3] += board[2][2]
                board[2][2] = 0
                moved = True
            if board[3][3] == 0 and board[3][2] != 0:  
                board[3][3] += board[3][2]
                board[3][2] = 0
                moved = True  
            if moved == False:
                print('Please choose a different direction')
                move(board)
            if moved == True:
                continue
                        
#bubble, selection, insertion, bogo sorts
#counting sort and radix sort
#time complexity (big O, theta, omega, running time function)
#linked lists
#binary search
#stack and queue




# priting to debug and use try/except
#what experiences they would look at ideal candidate

#Check if you won or lost
def win_check(board):
    
    for i in range(4):
        for j in range(4):
            if board[i][j] == 2048:
                return 'You won!'
    for i in range(4):
        for j in range(4):
            if board[i][j] == 0:
                return 'game not over'
    for i in range(3):
        for j in range(3):
            if board[i][j] == board[i+1][j] or board[i][j] == board[i][j+1]:
                return 'game not over'
            
    for j in range(3):
        if board[3][j] == board[3][j+1]:
            return 'game not over'
    for i in range(3):
        if board[i][3] == board[i+1][3]:
            return 'game not over'
        
    else:
        return 'You lost!'
```
